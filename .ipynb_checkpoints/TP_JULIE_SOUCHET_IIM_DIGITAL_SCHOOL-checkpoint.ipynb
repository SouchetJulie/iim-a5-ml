```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Preprocessing

- Check if missing values

- Check if duplicates (if yes, delete duplicated rows)

```python
# import data
train_df = pd.read_csv('./data/train.csv')
train_df.head()
```

```python
# check missing values
train_df.isnull().sum().sort_values(ascending=False)
# there seems to be no missing values
```

```python
# check duplicates (column id excluded because it is supposed to be unique)
print(train_df.duplicated(subset=train_df.columns[1:]).sum())
# remove the 304 duplicated rows
train_df.drop_duplicates(subset=train_df.columns[1:], inplace=True)
```
